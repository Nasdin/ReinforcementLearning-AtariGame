```python
import grab
```

```python
catalog_url = "https://en.wikipedia.org/wiki/Lists_of_musicians"
```

```python
from grab import Grab
g = Grab(log_file='grab.html')
```

```python
doc = g.go(catalog_url)
```

```python


lists = doc.select("//a[starts-with(text(),'List of')]")
links = map( lambda ls: "http://en.wikipedia.org"+ ls.node().attrib["href"], lists)
```

```python
from collections import defaultdict
person_links = []
person_categories = defaultdict(list)
for i, link in enumerate(links):
    doc_list = g.go(link)
    hrefs = doc_list.select("//a")

    for href in hrefs:
        attrs = href.node().attrib
        if 'href' not in attrs:
            continue
        if not attrs['href'].startswith("/wiki/"):
            continue
        person_url = "https://en.wikipedia.org"+attrs['href']
        person_links.append(person_url)
        person_categories[person_url].append(link)
    if i%10 ==0:
        print i, '/',len(links)
```

```python

def get_infobox_attrs(gr,url,table_name = "infobox"):

    doc = gr.go(url)
    table = doc.select("//table[starts-with(@class, '"+table_name+"')]")[0]
    rows = table.select("tr")
    
    attrs = {}
    for row in rows:
        try:
            key = row.select('th').text()
            val = row.select('td').text()
            if len(key) > 30:
                continue
            attrs[key]= val
        except:pass
    return attrs
```

```python
import joblib
import json


def _scrape_batch(urls,list_labels,output_fname = "./musicians0.json",logfile = "grab0.html"):
    
    gr = Grab(logfile)

    url_attrs = {}

    for i,(page_url,list_label) in enumerate(zip(urls,list_labels)):

        try:
            attrs = get_infobox_attrs(gr,page_url)
            attrs["_page_url"] = page_url
            attrs["_list"] = list_label
            
            url_attrs[page_url] = attrs
            
            #print "succeeded for",page_url
        except:
            #print "failed for", page_url
            pass
        if i%500==0:
            print i,'/',len(urls)
        
        
    with open(output_fname,'w') as fout:
        json.dump(url_attrs,fout)
    
    
    return True
```

```python
n_batches = 32

tasks = [ 
    joblib.delayed(_scrape_batch)(person_links[batch_i::n_batches],
                person_categories[batch_i::n_batches],
                "grab_musicians%i.json"%batch_i,
                "grab_log%i.html"%batch_i,)
 for batch_i in range(n_batches)]

res = joblib.Parallel(n_jobs=4)(tasks)
```
